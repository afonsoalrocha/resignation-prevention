```python
employee_df = pd.read_csv("HR_comma_sep.csv")
```

```python
left_df = employee_df[employee_df['left'] == 1]
stayed_df = employee_df[employee_df['left'] == 0]
```

```python
print(f"total: {len(employee_df)}")
print(f"left: {len(left_df)}")
print(f"left_percent: {len(left_df)/len(employee_df)}")
print(f"stayed: {len(stayed_df)}")
print(f"stayed_percent: {len(stayed_df)/len(employee_df)}")
```

```python
correlations = employee_df.corr()
f, ax = plt.subplots(figsize = (20,20))
sns.heatmap(correlations, annot=True);
```
